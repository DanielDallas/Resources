### Plotting two histograms on the same plot

```python
samples_1 = np.random.normal(loc=1, scale=.5, size=10000)
samples_2 = np.random.standard_t(df=10, size=10000)
bins = np.linspace(-3, 3, 50)

# set an alpha and using the same bins - this is to make the color transparent

plt.hist(samples_1, bins=bins, alpha=0.5, label='samples 1')
plt.hist(samples_2, bins=bins, alpha=0.5, label='samples 2')
plt.legend(loc='upper left')
plt.show()
```

### Scatter Plots

```python
plt.scatter(samples_1, samples_2, alpha=0.1)
plt.show()
```

## Histograms, Binnings and Density

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
```

```python
data = np.random.randn(1000)
data[:5]
```

```python
_ = plt.hist(data)
```

```python
_ = plt.hist(data, bins=30, normed=True, alpha=0.5, histtype='stepfilled', color='red')
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.3, normed=True, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

## Two Dimensional Histograms and Binnings

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
x
```

```python
y
```

```python
x.shape
```

```python
plt.hist2d(x, y, bins=10, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

```python
counts, xedges, yedges = np.histogram2d(x, y, bins=30)
counts, xedges, yedges
```

```python
counts.shape
```

<br><br>

### plt.hexbin: Hexagonal binnings
The two-dimensional histogram creates a tesselation of squares across the axes. Another natural shape for such a tesselation is the regular hexagon.

```python
plt.hexbin(x, y, gridsize=10, cmap='Blues')
cb = plt.colorbar(label='count in bin')
```

```python
import numpy as np
from matplotlib import pyplot as plt

scatter_x = np.array([1,2,3,4,5])
scatter_y = np.array([5,4,3,2,1])
group = np.array([1,3,2,1,3])
cdict = {1: 'red', 2: 'blue', 3: 'green'}

fig, ax = plt.subplots()
for g in np.unique(group):
    ix = np.where(group == g)
    ax.scatter(scatter_x[ix], scatter_y[ix], c = cdict[g], label = g, s = 100)
ax.legend()
plt.show()
```
